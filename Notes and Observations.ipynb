# Notes and Observations

Here is a small compilation of a few characteristics I found interesting when parsing through the data. I'm not placing value judgement on these observations, so feel free to discuss with me at the Espresso Afficionados Discord Server.

## Different coffee, same grinder

Shown below are 6 different coffees on the same grinder, at the same setting. The grinder used was a Niche Zero, and the coffees were as following:
1. Blackbird Blend by Fat Poppy Coffee
2. El Salvadore Honey Process by Fat Poppy Coffee
3. Hollywood Blend by Fat Poppy Coffee
4. YirgZero by Five Senses Coffee
5. Marimbus Espresso Blend by Market Lane
6. Final Frontier by Proud Mary

The Niche Zero has 63mm Mazzer Kony conical espresso burrs.

```python
#Load files 
df = pd.read_csv("niche_zero_fatpoppy_blackbird.csv")
df["DIAMETER"] = (2*np.sqrt(df["LONG_AXIS"]*df["SHORT_AXIS"]))/df["PIXEL_SCALE"]
df2 = pd.read_csv("niche_zero_fatpoppy_elsalvadore.csv")
df2["DIAMETER"] = (2*np.sqrt(df2["LONG_AXIS"]*df2["SHORT_AXIS"]))/df2["PIXEL_SCALE"]
df3 = pd.read_csv("niche_zero_fatpoppy_hollywood.csv")
df3["DIAMETER"] = (2*np.sqrt(df3["LONG_AXIS"]*df3["SHORT_AXIS"]))/df3["PIXEL_SCALE"]
df4 = pd.read_csv("niche_zero_fivesenses_yirg.csv")
df4["DIAMETER"] = (2*np.sqrt(df4["LONG_AXIS"]*df4["SHORT_AXIS"]))/df4["PIXEL_SCALE"]
df5 = pd.read_csv("niche_zero_marketlane_brazil_natural.csv")
df5["DIAMETER"] = (2*np.sqrt(df5["LONG_AXIS"]*df5["SHORT_AXIS"]))/df5["PIXEL_SCALE"]
df6 = pd.read_csv("niche_zero_proudmary_guji.csv")
df6["DIAMETER"] = (2*np.sqrt(df6["LONG_AXIS"]*df6["SHORT_AXIS"]))/df6["PIXEL_SCALE"]
```

```python
sns.set()
sns.set(rc={'figure.figsize':(8,6)})
plot = sns.kdeplot(data=df, x=df["DIAMETER"], 
             weights=df["VOLUME"],
             log_scale=True,
             label="Blackbird",
            common_norm=True
            )
plot2 = sns.kdeplot(data=df2, x=df2["DIAMETER"], 
             weights=df2["VOLUME"],
             log_scale=True,
             label="El Salvadore"
            )
plot3 = sns.kdeplot(data=df3, x=df3["DIAMETER"], 
             weights=df3["VOLUME"],
             log_scale=True,
             label="Hollywood"
            )
plot4 = sns.kdeplot(data=df4, x=df4["DIAMETER"], 
             weights=df4["VOLUME"],
             log_scale=True,
             label="YirgZero"
            )
plot5 = sns.kdeplot(data=df5, x=df5["DIAMETER"], 
             weights=df5["VOLUME"],
             log_scale=True,
             label="Marimbus"
            )
plot6 = sns.kdeplot(data=df6, x=df6["DIAMETER"], 
             weights=df6["VOLUME"],
             log_scale=True,
             label="Final Frontier"
            )
plot.set(xlabel="Diameter (mm)",ylabel="Mass Fraction (Volume Weighted)")
plot.legend()
```

**Note**: A KDE plot was used as opposed to a histogram for clarity of presentation for numerous distributions. As such, because the frequency count is weighted by mass, the y-axis fraction representation is relative, and will not sum to 1. Plots/Histograms used elsewhere will be normalized such that bins add up to 1.

## Changing grind settings

Using the same coffee, distributions were recorded on a Comandante C40 using the same coffee at 16, 20, and 24 clicks. According to documentation, each click corresponds to 30um of burr movement. The coffee used was a Kenya Nyeri from S&W Roasting.

```python
#load data
df = pd.read_csv("16_clicks.csv")
df["DIAMETER"] = (2*np.sqrt(df["LONG_AXIS"]*df["SHORT_AXIS"]))/df["PIXEL_SCALE"]
df2 = pd.read_csv("20_clicks.csv")
df2["DIAMETER"] = (2*np.sqrt(df2["LONG_AXIS"]*df2["SHORT_AXIS"]))/df2["PIXEL_SCALE"]
df3 = pd.read_csv("24_clicks.csv")
df3["DIAMETER"] = (2*np.sqrt(df3["LONG_AXIS"]*df3["SHORT_AXIS"]))/df3["PIXEL_SCALE"]
```

```python
plot = sns.histplot(df, x=df["DIAMETER"], 
             weights=df["VOLUME"],
             stat="probability",
             log_scale=True,kde=True,
             label="16 clicks",
             binwidth=0.06,
             element="step"
            )
plot.set(xlabel="Diameter (mm) (log-scale)",ylabel="Mass Fraction")
plot2 = sns.histplot(data=df2, x=df2["DIAMETER"], 
            weights=df2["VOLUME"],
            stat="probability",
            log_scale=True,kde=True,
            label="20 clicks",
            binwidth=0.06,
            color = 'r',
            element="step"
           )
plot3 = sns.histplot(data=df3, x=df3["DIAMETER"], 
            weights=df3["VOLUME"],
            stat="probability",
            log_scale=True,kde=True,
            label="24 clicks",
            binwidth=0.06,
            color = 'g',
            element="step"
         )
plot.legend()
```

An interesting coincidence is that 20 clicks was the grind setting I dialed to for a 18:300 v60 pourover brew at 21% EY.

## Shim alignment to a clean marker test

The sample below was a Fellow Ode with an burr swap to SSP Unimodal silver-knight, and was shim aligned to a clean [marker test](https://www.youtube.com/watch?v=8BCvoo5Sm2c). The marker test is semi-contentious, due to it not representing the full-tolerance stack and dynamic alignment with coffee in the burrset. Alignment is more of a concern with espresso grinding, while egregious misalignment should only have a noticeable effect with brewing grind sizes.

```python
df = pd.read_csv("fellow_ode_SSPunimodal_unknown_sample_pre_align.csv")
df["DIAMETER"] = (2*np.sqrt(df["LONG_AXIS"]*df["SHORT_AXIS"]))/df["PIXEL_SCALE"]
df2 = pd.read_csv("fellow_ode_SSPunimodal_unknown_sample_shim_align.csv")
df2["DIAMETER"] = (2*np.sqrt(df2["LONG_AXIS"]*df2["SHORT_AXIS"]))/df2["PIXEL_SCALE"]
```

```python
plot = sns.histplot(df, x=df["DIAMETER"], 
             weights=df["VOLUME"],
             stat="probability",
             log_scale=True,kde=True,
             label="SSP Ode Pre-Alignment",
             binwidth=0.06,
             element="step"
            )
plot.set(xlabel="Diameter (mm) (log-scale)",ylabel="Mass Fraction")
plot2 = sns.histplot(data=df2, x=df2["DIAMETER"], 
            weights=df2["VOLUME"],
            stat="probability",
            log_scale=True,kde=True,
            label="SSP Ode Shim-aligned",
            binwidth=0.06,
            color = 'r',
            element="step"
           )
plot.legend(loc="upper left")
```

## Burr seasoning

Of note; users often report having to grind a bit finer for the same shot times after seasoning. We also see a larger gain in grind quality than expected.

## RPM effects at filter coffee settings

For this data, a Lagom P64 with SSP Unimodal burrs was used, with Black and White Coffee's The Classic. The RPM's were as following:
1. Low: 200rpm
2. Med: 800rpm
3. High: 1400rpm

```python
df = pd.read_csv("bw_classic_low_rpm_data.csv")
df["DIAMETER"] = (2*np.sqrt(df["LONG_AXIS"]*df["SHORT_AXIS"]))/df["PIXEL_SCALE"]
df2 = pd.read_csv("bw_classic_med_rpm_data.csv")
df2["DIAMETER"] = (2*np.sqrt(df2["LONG_AXIS"]*df2["SHORT_AXIS"]))/df2["PIXEL_SCALE"]
df3 = pd.read_csv("bw_classic_high_rpm_data.csv")
df3["DIAMETER"] = (2*np.sqrt(df3["LONG_AXIS"]*df3["SHORT_AXIS"]))/df3["PIXEL_SCALE"]
```

```python
plot = sns.histplot(df, x=df["DIAMETER"], 
             weights=df["VOLUME"],
             stat="probability",
             log_scale=True,kde=True,
             label="200 RPM",
             binwidth=0.06,
             element="step"
            )
plot.set(xlabel="Diameter (mm) (log-scale)",ylabel="Mass Fraction")
plot2 = sns.histplot(data=df2, x=df2["DIAMETER"], 
            weights=df2["VOLUME"],
            stat="probability",
            log_scale=True,kde=True,
            label="800 RPM",
            binwidth=0.06,
            color = 'r',
            element="step"
           )
plot3 = sns.histplot(data=df3, x=df3["DIAMETER"], 
            weights=df3["VOLUME"],
            stat="probability",
            log_scale=True,kde=True,
            label="1400 RPM",
            binwidth=0.06,
            color = 'g',
            element="step"
         )
plot.legend(loc="upper left")
```

There is debate about whether changing the grind RPM changes the distribution shape itself, or changes the relative grind setting. At filter settings, it appears not to have discernable effects. However, most users use it for espresso.
